```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)

df = pd.read_csv("QQQ_base-QQQ_base.csv")
df = df.drop(df.columns[[58, 59]], axis=1)

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").set_index("Date")

df.head()
```

```python
df.isna().sum()
```

```python
fig, axes = plt.subplots(3, 1, figsize = (12,10), sharex=True)
df["QQQ_close"].plot(ax=axes[0])
axes[0].set_title("QQQ_close")

df["VIX"].plot(ax=axes[1])
axes[1].set_title("VIX")

df["QQQ_volume"].plot(ax=axes[2])
axes[2].set_title("QQQ_volume")

plt.tight_layout()
```

```python
ret = df["QQQ_today_close_to_tmrw_close_return"]
baseline_ret = df["baseline_strategy_return"]

fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

ret.plot(ax=axes[0])
axes[0].set_title("QQQ TT+1 Returns")

baseline_ret.plot(ax=axes[1])
axes[1].set_title("Baseline Strategy Returns")

plt.tight_layout()
```

```python
bh_equity = (1 + ret).cumprod()
baseline_equity = (1 + baseline_ret).cumprod()

plt.figure(figsize=(12, 5))
bh_equity.plot(label="QQQ Buy-and-Hold")
baseline_equity.plot(label="Baseline Strategy")
plt.legend()
plt.title("Cumulative Performance")
plt.show()
```

```python
lt_cols = [c for c in df.columns if c.startswith("LT_SIG")]
st_cols = [c for c in df.columns if c.startswith("ST_SIG")]

df["n_LT_on"] = df[lt_cols].sum(axis=1)
df["n_ST_on"] = df[st_cols].sum(axis=1)
df["n_total_on"] = df["n_LT_on"] + df["n_ST_on"]

cols = ["n_LT_on", "n_ST_on", "n_total_on"]

df[cols].describe(percentiles=np.arange(0.05, 1.00, 0.05))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

df["n_LT_on"].plot(kind="hist", bins=range(0, 12), ax=axes[0])
axes[0].set_title("# Long-Term Signals Active")

df["n_ST_on"].plot(kind="hist", bins=range(0, 45, 2), ax=axes[1])
axes[1].set_title("# Short-Term Signals Active")

df["n_total_on"].plot(kind="hist", bins=range(0, 55, 2), ax=axes[2])
axes[2].set_title("# Total Signals Active")

plt.tight_layout()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 9), sharex=True)

df["n_LT_on"].plot(ax=axes[0])
axes[0].set_title("n_LT_on over time")

df["n_ST_on"].plot(ax=axes[1])
axes[1].set_title("n_ST_on over time")

df["n_total_on"].plot(ax=axes[2])
axes[2].set_title("Total signals over time")
```

Got a good idea of the signal activity. This can be used while tuning parameters for optimization.

```python
corr = df[[
    "n_LT_on", "n_ST_on", "n_total_on",
    "VIX",
    "baseline_target_exposure_calculated_at_todays_close",
    "QQQ_today_close_to_tmrw_close_return",
    "baseline_strategy_return"
]].corr()

corr
```

From this, I can conclude that baseline exposure is mainly driven by the signals. 
The negative correlation between the baseline exposure and VIX shows that volatility interacts with the strategy, but only weakly.

Lastly, short term signals are more active than long term signals during vol spikes.

```python
df["LT_any"] = (df["n_LT_on"] > 0).astype(int)

df.groupby("LT_any")["QQQ_today_close_to_tmrw_close_return"].mean()
df.groupby("LT_any")["baseline_strategy_return"].mean()
```

```python
df["LT_bucket"] = pd.cut(df["n_LT_on"], bins=[-0.5, 0.5, 1.5, 3.5, 10.5],
                         labels=["0", "1", "2-3", "4+"])

df.groupby("LT_bucket")[["QQQ_today_close_to_tmrw_close_return",
                         "baseline_strategy_return"]].mean()
```

```python
df.groupby("LT_bucket")["baseline_target_exposure_calculated_at_todays_close"].mean()
```

```python
df["VIX_bucket"] = pd.qcut(df["VIX"], q=3, labels=["low", "med", "high"])

df.groupby("VIX_bucket")[["QQQ_today_close_to_tmrw_close_return",
                          "baseline_strategy_return"]].mean()
df.groupby("VIX_bucket")["baseline_target_exposure_calculated_at_todays_close"].mean()
```
